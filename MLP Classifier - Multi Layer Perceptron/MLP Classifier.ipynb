```python
import pandas as pd
import string
```

```python
df = pd.read_csv('final_all_names_code.csv')
df.head()
```

```python
columns = list(string.ascii_lowercase) + ['Country_code','Country']
```

```python
newdf = pd.DataFrame(columns=columns)
newdf
```

```python
hasharrays = []
for ind in range(10000):
    name = df['Name'][ind]
    nlst = str(name).split()
    lastn = nlst[-1].lower()
    chars = dict.fromkeys(columns,0)
    for i in lastn:
        chars[i] = 1
    chars['Country_code'] = df['Country_code'][ind]
    chars['Country'] = df['Country'][ind]
    hasharrays.append(chars)
aux_df = pd.DataFrame(hasharrays)
newdf = pd.concat([newdf,aux_df],ignore_index=True)
```

```python
newdf.head()
```

```python
newdf.drop(['.'],axis=1,inplace=True)
newdf
```

```python
X = newdf[list(string.ascii_lowercase)]
Y = newdf['Country']
X
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
X = StandardScaler().fit_transform(X)
Y = LabelEncoder().fit_transform(Y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(100,50,30),solver='adam',activation='relu',random_state=42,verbose=True,
learning_rate_init=0.01,n_iter_no_change=30,max_iter=200)

mlp.fit(xtrain,ytrain)

y_pred = mlp.predict(xtest)
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_pred)
```
